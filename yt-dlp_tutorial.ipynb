# How to download a video from Youtube using yt-dlp


# Step 1 - Install yt-dlp
**Windows**
1. Download the latest yt-dlp.exe from the official GitHub releases page: https://github.com/yt-dlp/yt-dlp/releases
2. Put it somewhere handy (e.g., C:\yt-dlp\yt-dlp.exe).
3. (Optional) Add that folder to your **PATH** so you can just type yt-dlp from any command prompt.

**MacOS**
1. Install Homebrew if you do not have it already
Open **terminal** and paste:

````python
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
````

To check if Homebrew is already installed, type: 

````python
brew --version
````

2. Install yt-dlp

````python
brew install yt-dlp
````

# Step 2 - Install FFMEG
**Windows**
1. Download from FFmpeg.org https://ffmpeg.org/download.html
2. Extract, and add bin folder to **PATH**

**macOS**
Open **terminal** and type
```console
brew install ffmpeg
```
To verify the installation, you can type
```console
yt-dlp --version
```
Tada! You're done with MacOs.


# Step 3 **Windows ONLY**
Add yt-dlp and ffmeg to path so you can access them better
1.	Press Win + R, type sysdm.cpl, press Enter.
2.	Go to Advanced → Environment Variables.
3.	Under System variables, find Path → Edit → New.
o	Add: C:\yt-dlp\
o	Add: C:\ffmpeg\bin\
4.	Click OK → OK → OK.
Now you can open a new Command Prompt and just type:
```console
yt-dlp --version
```
If it shows a version number, you're good to go!

# Step 4 - Try it out!
Type this in your console

````python
yt-dlp -f "bestvideo+bestaudio" --merge-output-format mp4 -o "%(title)s.mp4" VIDEO_URL
````

Keep it updated with:

````python
yt-dlp -U
````

# Troubleshooting yt-dlp (if using public wifi)

````python
How to test if it’s the Wi-Fi

Try the same command at home or on mobile hotspot. If it works fine there, the campus network is the issue.

Run with verbose logging:
````

````python
yt-dlp -v URL
````

Look for “HTTP error”, “connection reset”, or “timeout”

## Things you can try on public wifi
1. Limit download speed (reduces strain on network, helps avoid triggering filters):

````python
yt-dlp --limit-rate 500K URL
````

2. Use a different format (lower quality → smaller size → less chance of timeout):

````python
yt-dlp -f "bv*[height<=480]+ba" --merge-output-format mp4 URL
````

3. Force single connection (instead of multiple fragments):

````python
yt-dlp --concurrent-fragments 1 URL
````
